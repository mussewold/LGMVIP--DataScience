# Image to Pencil Sketch with Python

### Importing the libraries

```python
import matplotlib.pyplot as plt
import cv2
```

###  Reading the 'Mesob's' image from file


```python
image=cv2.imread("mesob.jpg")
plt.imshow(image)
plt.show()
```

### Let's Convert the BGR to RGB image

```python
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

### Let's Convert the RGB image to Gray Scale

```python
gray_img=cv2.cvtColor(rgb_img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

### Inverting the gray scale image

```python
inverted_image= cv2.bitwise_not(gray_img)
plt.imshow(inverted_image)
```

### Blurring the image using Gaussian Function

```python
blur_img=cv2.GaussianBlur(inverted_image, (21, 21),0)
plt.imshow(blur_img)
```

### Mixing the Gray-Scale Image with Inverted Blurry Image

```python
invblur_img = cv2.bitwise_not(blur_img)
plt.imshow(invblur_img)
```

### Performing bit-wise division
The sketch can be generated by bit-wise division the grayscale image and the inverted-blurred image.

```python
sketch_img=cv2.divide(gray_img,invblur_img, scale=256.0)
```

### Saving the sketch image

```python
 cv2.imwrite('mesob.png', sketch_img)
```

### Displaying the original image and final sketch

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original Mesob Imgae', size=18)
plt.imshow(rgb_img)
plt.axis('off')
```

## What is Mesob by the way?? 

The Amharic word Mesob means 'breadbasket' and it's where Ethiopians keep our injera, our traditional flat bread on which we serve our beautiful stews, which are made with traditional herbs and spices.

```python

plt.figure(figsize=(14,8))
plt.title('Sketched Mesob Image', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
